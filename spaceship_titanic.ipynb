```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
df.head(3)
```

```python
df.drop(columns=['Name','Cabin','PassengerId'],inplace=True)
df = pd.get_dummies(df,columns=['HomePlanet','VIP','Destination','CryoSleep'],drop_first=True)
df.head(3)
```

```python
x = df.drop(columns=['Transported'])
y = df['Transported']
```

```python
X_train, y_train = x,y
```

```python
df.head(3)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
test_data = pd.read_csv('test.csv')
```

```python
test_data.head(3)
```

```python
test_data.drop(columns=['Name','Cabin','PassengerId'],inplace=True)
test_data = pd.get_dummies(test_data,columns=['HomePlanet','VIP','Destination','CryoSleep'],drop_first=True)
test_data.head(3)
```

```python
predictions = model.predict(test_data)
```

```python
test_data = pd.read_csv('test.csv')
```

```python
# Create a DataFrame for the submission
submission = pd.DataFrame({
    'PassengerId': test_data['PassengerId'],
    'Transported': predictions
})


submission['Transported'] = submission['Transported'].astype(bool)


submission.to_csv('/content/submission.csv', index=False)
```
